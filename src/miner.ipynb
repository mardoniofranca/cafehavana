```python
def tratar_line(line):
    line = line.replace("\t",";")
    line = line.replace('"','')
    line = line.replace('\n','')  
    line = line.replace('.','')
    line = line.replace(',','.')
    return line
```

```python
line_meses = ""
lista      = []
line_valor = ""
i = 0
f = open("../data/tab_cooxupe.txt", "r")
for x in f:
    if i == 0 :
        line_meses = x
    if i > 0 :  
        line_valor = x
        lista.append(tratar_line(line_valor))
    i = i +1 
```

```python
matriz = []
vetor  = []
data   = []
col_names =  ['data', 'valor']
df  = pd.DataFrame(columns = col_names)


for l in lista:
    vetor = l.split(';')
    j = 0
    for v in vetor:
        if j == 0:
            ano = v
        else:
            mes = j  
            if len(v) == 0:
                valor = 0
            else:
                valor = v
            data_valor = str(ano) + '_' + str(mes) + ';'+ str(valor)
            if j > 0 :
                print(data_valor) 
                values = data_valor.split(';')
                df = df.append({'data': values[0], 'valor': values[1]}, ignore_index=True)
                data.append(data_valor)   
        j = j + 1
```
